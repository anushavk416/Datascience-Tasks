#NETFLIX DATASET CLEANING

```python
#importing libraries
import numpy as np
import pandas as pd
```

```python
#reading data
df = pd.read_csv("/content/netflix_titles.csv")
df.head()
```

```python
df.info()
```

```python
# checking for duplicates
df.drop_duplicates(inplace = True)
df.shape
```

```python
# checking type
df['type'].value_counts()
df['rating'].unique()
```

```python
df.loc[5541, 'duration'] = df.loc[5541, 'rating']
df.loc[5794, 'duration'] = df.loc[5794, 'rating']
df.loc[5813, 'duration'] = df.loc[5813, 'rating']
```

```python
df['duration'].unique()
```

```python
df['duration'].isnull().sum()
```

```python
df.info()
```

as all titles and show id are available, it means that the shows are present on the platform but some do not have their details filled. because of which we can't drop the records. so, filling them with information is a better option.

```python
df['director'] = df['director'].fillna('unknown')
df['cast'] = df['cast'].fillna('unknown')
df['country'] = df['country'].fillna('unknown')
df['date_added'] = df['date_added'].fillna('unknown')
```

```python
df.info()
```

```python
# handling duration values in rating
df.loc[df['rating'].str.contains('min', na=False), 'rating'] = np.nan
df['rating'] = df['rating'].fillna('unknown')
df[df['rating'].str.contains('min')]
```
