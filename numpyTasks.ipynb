```python
import numpy as np
```

# Part 1: Creating and Reshaping Arrays


```python
# Creating arrays
# 1D array of 10 equally spaced values between 0 and 50
arr1 = np.linspace(0, 50, 10)
print("1D Array:")
print(arr1)

# 2D array (3x3) random integers between 5 and 20
arr2 = np.random.randint(5, 21, (3,3))
print("2D Array:")
print(arr2)
```

```python
# 2. Array Dimensions
print("Dimensions of arr1:", arr1.ndim)
print("Dimensions of arr2:", arr2.ndim)

print("Shape of arr1:", arr1.shape)
print("Shape of arr2:", arr2.shape)
```

```python
# 3. Reshaping
reshaped = arr1.reshape(2,5)
print("Reshaped Array (2x5):")
print(reshaped)

flattened = reshaped.flatten()
print("Flattened back to 1D:")
print(flattened)
```

# Part 2: Indexing and Slicing

```python
# 4. Indexing
# element at 2nd row, 3rd column
print("Element at 2nd row, 3rd column:")
print(arr2[1,2])

# first two rows and last two columns
print("First two rows and last two columns:")
print(arr2[:2, -2:])
```

```python
# 5. Slicing

# first four elements from flattened array
print("First four elements:")
print(flattened[:4])

# reverse the 1D array
print("Reversed 1D array:")
print(arr1[::-1])
```

# Part 3: Array Operations


```python
# 6. Create another 1D array
arr3 = np.random.randint(1,11,10)
print("New Array (1 to 10 integers):")
print(arr3)

# Addition
print("Addition:")
print(arr1 + arr3)

# Multiplication
print("Multiplication:")
print(arr1 * arr3)

# Subtract constant
print("Subtract 5:")
print(arr1 - 5)
```

# Part 4: Random Number Generation

```python
# random values between 0 and 1
rand1 = np.random.rand(5)
print("Random values (0 to 1):")
print(rand1)

# random values between -1 and 1
rand2 = np.random.uniform(-1,1,5)
print("Random values (-1 to 1):")
print(rand2)
```
