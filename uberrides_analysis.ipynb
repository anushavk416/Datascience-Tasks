```python
# import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# reading file
df=pd.read_csv('https://github.com/Taj1920/Data_Analysis/raw/refs/heads/main/Uber_Analysis/UberDataset.csv')
df.head()
```

```python
df.info()
```

```python
# last row can be removed
df = df.drop([1155])
df
```

```python
#Removing duplicate rows
df.drop_duplicates(inplace=True)
df.info()
```

```python
#Resetting the index
df.reset_index(inplace=True)
df.drop('index',axis=1,inplace=True)
df.head()
```

```python
#Handling nulls in PURPOSE
df['PURPOSE']=df['PURPOSE'].fillna('Unknown')
df.info()
```

```python
#handling Kar?chi value
df['START']=df['START'].str.replace('Kar?chi','Karchi')
df['STOP']=df['STOP'].str.replace('Kar?chi','Karchi')
```

```python
#changing to datetime datatype
df['START_DATE']=df['START_DATE'].str.replace('/','-')
df['END_DATE']=df['END_DATE'].str.replace('/','-')


df['START_DATE']=pd.to_datetime(df['START_DATE'])
df['END_DATE']=pd.to_datetime(df['END_DATE'])
df.info()
```

```python
# Putting the START_DATE to date and time column (Feature engineering

df['DATE']=pd.DatetimeIndex(df['START_DATE']).date
df['TIME']=pd.DatetimeIndex(df['START_DATE']).hour

# converting the time into four different categories i.e. Morning, Afternoon, Evening, Night
out=[]
for i in df['TIME']:
  if 0<=i<=10:
    out.append('Morning')
  elif 11<=i<=15:
    out.append('Afternoon')
  elif 16<=i<=19:
    out.append('Evening')
  elif 20<=i<=24:
    out.append('Night')
df['DAY-NIGHT']=out
df
```

```python
#creating month column
df['MONTH']=[i.month for i in df['START_DATE']]
df['MONTH']=df['MONTH'].map({1:'JAN',  2:'FEB',  3:'MAR',  4:'APR',  5:'MAY',  6:'JUN',  7:'JUL',  8:'AUG',  9:'SEP', 10:'OCT', 11:'NOV', 12:'DEC'})
df
```

```python
#creating day column
df['DAY']=df['START_DATE'].dt.weekday
df['DAY']=df['DAY'].map({0:'MON',1:'TUE',2:'WED',3:'THU',4:'FRI',5:'SAT',6:'SUN'})
df
```

# Data visualization

```python
# count based on cateogry
a = sns.histplot(x='CATEGORY',data=df,hue='CATEGORY',palette='dark:black')
def data_label(a):
  for i in a.containers:
    a.bar_label(i,label_type='center',color='white')
data_label(a)
```

Count of Business category is more

```python
# count based on PURPOSE
uber_palette = sns.color_palette(["black","gray"])
a=sns.countplot(x='PURPOSE',data=df,hue='CATEGORY',palette=uber_palette)
plt.xticks(rotation=90)
plt.show()
```

```python
# count based on DAY-NIGHT
a=sns.countplot(x='DAY-NIGHT',data=df,hue='DAY-NIGHT',palette='dark:black')
data_label(a)
plt.xticks(rotation=90)
plt.show()
```

```python
# count based on MONTH
plt.figure(figsize=(10,5))
a=sns.histplot(x='MONTH',data=df,hue='MONTH',palette='dark:black')
data_label(a)
plt.show()
```

```python
# count based on DAY
plt.figure(figsize=(10,5))
a=sns.countplot(x='DAY',data=df,hue='DAY',palette='dark:black')
data_label(a)
plt.show()
```

```python
# count based on MILES
plt.figure(figsize=(10,5))
sns.histplot(x='MILES',data=df[df['MILES']<100],kde=True,bins=50,color='black')
plt.show()
```

```python
# count based on START time
df1=pd.DataFrame(df['START'].value_counts())
plt.figure(figsize=(10,5))
a=sns.barplot(x='START',y='count',data=df1[df1['count']>30],hue='START',palette='dark:black')
data_label(a)
```
